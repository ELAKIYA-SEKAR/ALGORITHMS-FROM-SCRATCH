```python
import pandas as pd
import numpy as np
```

```python
dff=pd.read_csv("/content/toy_dataset.csv")
```

```python
data=dff.head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data=data[['Age','Income']]
df=pd.DataFrame(scaler.fit_transform(data),columns=data.columns, index=data.index)
```

```python
df
```

```python
x = np.array(df.Age)
y = np.array(df.Income)
```

```python
def error(y_pred):
  return np.mean(np.square(y-y_pred))
```

```python
m=1
c=1
learning_rate=0.001
epochs=1000
n=len(x)
error_vals=[]
```

```python
for i in range(epochs):
  y_pred=m*x+c
  error_vals.append(error(y_pred))
  de_dm= (-2/n)* np.dot(x , (y-y_pred))
  de_dc= (-2/n) * np.sum(y-y_pred)
  m = m - (de_dm * learning_rate)
  c = c - (de_dc * learning_rate)
```
